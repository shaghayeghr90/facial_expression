```python
def predict_expressions(path):
    total_num_images=0
    for filename in glob.glob(os.path.join(path, '*.jpg')):
        total_num_images=total_num_images+1
            
    faces=[]
    valid_frames=[]
    for filename in glob.glob(os.path.join(path, '*.png')):
        if 'landmark' not in filename:
            temp=filename.split('/')
            temp=temp[10].split(' ')[1].split('.')[0]
            valid_frames.append(int(temp))
            image = cv2.imread(filename)
            image=scipy.misc.imresize(image,size=(48,48))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            temp=np.expand_dims(image,2)
            image=np.concatenate((temp,temp),axis=2)
            image=np.concatenate((image,temp),axis=2)
            faces.append(image)
    X=np.array(faces) / 255.0
    frames=np.array(valid_frames)
    
    print('X: ',X.shape)   
    print('frames: ',frames.shape)
    predictions=model.predict(X, batch_size=32, verbose=0)
    print('predcitions: ',predictions.shape)
    return predictions,frames,total_num_images
```

```python
for name in player_names:
    predictions,frames,total_num_images=predict_expressions(path+name)
    with open('vgg_lstm_predictions/'+name+'.csv', 'w') as csvfile:
        fieldnames = ['Angry', 'Disgust','Fear','Happy','Sad','Surprise','Neutral']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()
        count=0
        for i in range(total_num_images):
            row={}
            if i+1 in frames:
                row['Angry']=predictions[count,0]
                row['Disgust']=predictions[count,1]
                row['Fear']=predictions[count,2]
                row['Happy']=predictions[count,3]
                row['Sad']=predictions[count,4]
                row['Surprise']=predictions[count,5]
                row['Neutral']=predictions[count,6]
                writer.writerow(row)
                count=count+1
            else:
                row['Angry']=-10
                row['Disgust']=-10
                row['Fear']=-10
                row['Happy']=-10
                row['Sad']=-10
                row['Surprise']=-10
                row['Neutral']=-10
                writer.writerow(row)
```
